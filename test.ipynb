```python
#code to extract facebook links from excel
#test_dataframe = pd.read_excel("EnvNP_SG.xlsx")
from urllib.parse import urlparse
import ast 
import numpy as np
def get_facebook_links(social_medias):
    social_medias = ast.literal_eval(social_medias)
    for social_media in social_medias:
        if "facebook" in social_media:
            return social_media
    return np.NaN

def extract_user(url):
    if url is np.NaN: return np.NaN
    parsed_url = urlparse(url)
    path_parts = parsed_url.path.split('/')
    # The user is the second part for the first type of URL and the third part for the second type of URL
    if path_parts[-1].strip() == "":
        user = path_parts[-2]
    elif len(path_parts) > 2 and path_parts[1] != "groups":
        user = path_parts[1]
    else:
        user = path_parts[-1]
    
    return user

countries = ["SG"] #Malaysia, Laos
dataframe_map = {}
for country in countries:
    print(country)
    test_dataframe = pd.read_excel(f"EnvNP_{country}.xlsx")
    test_dataframe["Facebook"] =  test_dataframe["Social Medias"].apply(get_facebook_links)
    #given a link in this format https://www.facebook.com/GreenBrunei/	, i want to extract the user which is GreenBrunei

    test_dataframe['Facebook_User'] = test_dataframe['Facebook'].apply(extract_user)    
    test_dataframe["Country"] = country
    #create a dataframe map
    dataframe_map[country] = test_dataframe
```

```python
#IP Rotator
import requests
import random
from selenium.common.exceptions import TimeoutException, WebDriverException
from facebook_page_scraper import Facebook_scraper

proxies = [
    '72.10.160.91:8080',
    '20.37.207.8:8080',
    '72.10.164.178:8080',
    "50.174.145.12:8080",
    "162.223.94.164:8080",
    "38.54.79.150:8080",
    "102.130.125.86:8080",
    "20.247.228.80:8080",
    "170.64.222.86:8080",
    "128.199.202.122:8080",
    "50.230.222.202:8080"
]

def make_request(row, proxies=proxies):
    # Select a random proxy
    proxy = random.choice(proxies)
    url = row['Facebook']
    page_name = row['Facebook_User']
    if pd.isna(page_name):
        return pd.DataFrame({"Page_Name": page_name, "PostIDs" : [], "Comments": [], "Content": [], "Post_URL": [], "Images": [], "Date": [], "Reactions": [], "Reaction_Count": [], "Shares": [], "video": []})
    
    try:
        posts_count = 100
        browser = "firefox"
        #proxy = f"137.132.26.16:{port}" #if proxy requires authentication then user:password@IP:PORT
        timeout = 600 #600 seconds
        headless = False
        meta_ai = Facebook_scraper(page_name, posts_count, browser, proxy=proxy, timeout=timeout, headless=headless)

        json_data = meta_ai.scrap_to_json()
        test_data = ast.literal_eval(json_data)

        #preparing the dataset
        test_frame = pd.DataFrame({"Page_Name": page_name, "PostIDs" : test_data.keys(), "Value_Column": test_data.values()})
        test_frame['Comments'] = test_frame['Value_Column'].apply(lambda x: x['comments'])
        test_frame['Content'] = test_frame['Value_Column'].apply(lambda x: x['content'])

        test_frame['Post_URL'] = test_frame['Value_Column'].apply(lambda x: x['post_url'])
        test_frame['Images'] = test_frame['Value_Column'].apply(lambda x: x['image'])
        test_frame['Date'] = test_frame['Value_Column'].apply(lambda x: x['posted_on'])
        test_frame['Reactions'] = test_frame['Value_Column'].apply(lambda x: x['reactions'])
        test_frame['Reaction_Count'] = test_frame['Value_Column'].apply(lambda x: x['reaction_count'])
        test_frame['Shares'] = test_frame['Value_Column'].apply(lambda x: x['shares'])
        test_frame['video'] = test_frame['Value_Column'].apply(lambda x: x['video'])
        test_frame.drop(columns=['Value_Column'], inplace=True)
    except (SystemExit, Exception) as e:
        print(f"Timeout exception for {page_name}")
        test_frame = pd.DataFrame({"Page_Name": page_name, "PostIDs" : [], "Comments": [], "Content": [], "Post_URL": [], "Images": [], "Date": [], "Reactions": [], "Reaction_Count": [], "Shares": [], "video": []})
    return test_frame



for country in countries:
    result_frames = []
    country_data = dataframe_map[country]
    for index, row in country_data.iterrows():
        result = make_request(row)
        result_frames.append(result)
    country_df = pd.concat(result_frames)
    country_df.rename(columns={"Page_Name": "Facebook_User"}, inplace=True)
    country_df = pd.merge(country_data, country_df, on='Facebook_User', how='left')
    dataframe_map[country] = country_df
    country_df.to_excel(f"EnvNP2_{country}_Facebook.xlsx", index=False) #saving so don't have to run iteration again if lost timeout


    
```
